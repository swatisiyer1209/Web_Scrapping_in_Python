```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.scrapethissite.com/pages/forms/'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html')
print(soup)
```

```python
headers=soup.find_all('th')
table_headers=[header.text.strip() for header in headers]
print(table_headers)
```

```python
import pandas as pd
df=pd.DataFrame(columns=table_headers)
df
```

```python
for i in range(1,25):
    url='https://www.scrapethissite.com/pages/forms/?page_num='+str(i)
    page=requests.get(url)
    soup=BeautifulSoup(page.text,'html')
    rows=soup.find_all('tr')[1:]
    for row in rows:
        dataset=row.find_all('td')
        individual_data=[data.text.strip() for data in dataset]
        length=len(df)
        df.loc[length]=individual_data
df
```
